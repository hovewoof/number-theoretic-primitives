```python
def binary_euclid(num1, num2, need_to_print: bool):
    if need_to_print:
        print("Extended Binary Euclid Algorithm")
    a, b = num1, num2
    d = 1
    while a & 1 == 0 and b & 1 == 0:
        a = a >> 1
        b = b >> 1
        d = d * 2
    r1, r2, x1, x2 = a, b, 1, 0
    y1, y2 = 0, 1
    i = 2
    if need_to_print:
        print(f"\tStep: {0}\n\t\tr: {r1};\n\t\tx: {x1};\n\t\ty: {y1};")
    if need_to_print:
        print(f"\tStep: {1}\n\t\tr: {r2};\n\t\tx: {x2};\n\t\ty: {y2};")
    while r1 != 0:
        while r1 & 1 == 0:
            r1 = r1 >> 1
            if x1 & 1 == 0 and y1 & 1 == 0:
                x1, y1 = x1 >> 1, y1 >> 1
            else:
                x1, y1 = (x1 + b) >> 1, (y1 - a) >> 1
        while r2 & 1 == 0:
            r2 = r2 >> 1
            if x2 & 1 == 0 and y2 & 1 == 0:
                x2, y2 = x2 >> 1, y2 >> 1
            else:
                x2, y2 = (x2 + b) >> 1, (y2 - a) >> 1
        if abs(r1) >= abs(r2):
            r1, x1, y1 = r1 - r2, x1 - x2, y1 - y2
        else:
            r2, x2, y2 = r2 - r1, x2 - x1, y2 - y1
        if need_to_print:
            print(f"\tStep: {i}\n\t\tr: {r2*d};\n\t\tx: {x2};\n\t\ty: {y2};")
        i += 1
    if (x2 * num1 + y2 * num2) == r2 * d: 
        if need_to_print:
            print(f"\tGCD({num1}; {num2}) = {r2 * d * g};\n\t\tx = {x2};\n\t\ty = {y2}.")
    else:  
        if need_to_print:
            print("\tAlgorithm error.")
```

```python
def truncated_euclid(num1, num2, need_to_print: bool):
    if need_to_print:
        print("Truncated Extended Euclid Algorithm")
    i = 2
    r1, r2, x1, x2 = num1, num2, 1, 0
    y1, y2, q1, q2 =  0, 1, 0, num1 // num2
    if need_to_print:
        print(f"\tStep: {0}\n\t\tr: {r1};\n\t\tx: {x1};\n\t\ty: {y1};")
    if need_to_print:
        print(f"\tStep: {1}\n\t\tr: {r2};\n\t\tx: {x2};\n\t\ty: {y2};\n\t\tq: {q2};")
    while r2 != 0:
        q1 = r1 // r2
        r1, r2 = r2, r1 % r2
        if r2 != 0:
            q2 = r1 // r2
        x1, x2 = x2, x1 - q1 * x2
        y1, y2 = y2, y1 - q1 * y2
        if r2 > r1 / 2: 
            r2 = r1 - r2
            x2, y2 = x1 - x2, y1 - y2
        if need_to_print:
            print(f"\tStep: {i}\n\t\tr: {r2};\n\t\tx: {x2};\n\t\ty: {y2};\n\t\tq: {q2};")
        i += 1
    if (x1 * num1 + y1 * num2) == r1: 
        if need_to_print:
            print(f"\tGCD({num1}; {num2}) = {r1 * g};\n\t\tx = {x1};\n\t\ty = {y1}.")
    else: 
        if need_to_print:
            print("\tAlgorithm error.")
```
